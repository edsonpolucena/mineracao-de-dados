```python
"""
Comprometimento de renda das famílias – Tendência
-------------------------------------------------
Série SGS 29265 (Banco Central do Brasil).

• Baixa a série diretamente da API do BCB.
• Contorna o limite de 10 anos por requisição (após mar/2025).
• Calcula inclinação via regressão linear e teste de Mann-Kendall.
• Mostra estatísticas e salva/mostra gráfico.

Autor: Edson/Wuelliton
Data: 2025-05-25
"""
import os
from datetime import datetime, timedelta

import pandas as pd
import matplotlib.pyplot as plt
import requests
from sklearn.linear_model import LinearRegression
from scipy.stats import kendalltau

# ---------------- Configurações ----------------
SERIES_ID = 29265        # Comprometimento de renda – sem ajuste
DATA_INICIAL = "01/03/2005"
HOJE = datetime.today().strftime("%d/%m/%Y")
OUT_DIR = "outputs"
os.makedirs(OUT_DIR, exist_ok=True)


# ---------------- Funções utilitárias ----------------
def _fetch_chunk(series_id: int, start: str, end: str) -> pd.DataFrame:
    """Baixa até 10 anos de dados da série `series_id`."""
    url = (
        f"https://api.bcb.gov.br/dados/serie/bcdata.sgs.{series_id}/dados"
        f"?formato=json&dataInicial={start}&dataFinal={end}"
    )
    r = requests.get(url, timeout=30)
    r.raise_for_status()
    df = pd.read_json(r.text, dtype={"valor": "float"})
    df["data"] = pd.to_datetime(df["data"], format="%d/%m/%Y")
    return df


def fetch_series(series_id: int, start: str, end: str) -> pd.DataFrame:
    """Contorna o limite de 10 anos fazendo requisições em fatias."""
    start_dt = datetime.strptime(start, "%d/%m/%Y")
    end_dt = datetime.strptime(end, "%d/%m/%Y")
    dfs = []
    chunk_start = start_dt
    while chunk_start <= end_dt:
        # define chunk_end = chunk_start + 9 anos (máx.)
        try:
            chunk_end = chunk_start.replace(year=chunk_start.year + 9)
        except ValueError:                 # 29/02 em ano bissexto
            chunk_end = chunk_start + timedelta(days=365 * 9 + 2)
        if chunk_end > end_dt:
            chunk_end = end_dt

        dfs.append(
            _fetch_chunk(
                series_id,
                chunk_start.strftime("%d/%m/%Y"),
                chunk_end.strftime("%d/%m/%Y"),
            )
        )
        chunk_start = chunk_end + timedelta(days=1)
    return pd.concat(dfs).sort_values("data").reset_index(drop=True)


def mann_kendall(series: pd.Series):
    """Teste de Mann-Kendall (τ, p)."""
    tau, p = kendalltau(range(len(series)), series)
    return tau, p


def linear_slope(series: pd.Series):
    """Inclinação anualizada via regressão linear simples."""
    x = (series.index.astype(int) // 10 ** 9).values.reshape(-1, 1)  # segundos
    y = series.values.reshape(-1, 1)
    model = LinearRegression().fit(x, y)
    # coef → por segundo → por ano
    return model.coef_[0][0] * 60 * 60 * 24 * 365


# ---------------- Pipeline ----------------
def main():
    print("Baixando série SGS 29265…")
    df = fetch_series(SERIES_ID, DATA_INICIAL, HOJE)
    serie = df.set_index("data")["valor"]

    # Estatísticas
    slope = linear_slope(serie)
    tau, p = mann_kendall(serie)
    tendencia = (
        "CRESCENTE" if slope > 0 and p < 0.05
        else "DECRESCENTE" if slope < 0 and p < 0.05
        else "ESTÁVEL"
    )

    print("\n=== Comprometimento da renda ===")
    print(f"Slope (regressão linear): {slope:.4f} pp/ano")
    print(f"Mann-Kendall τ / p-valor : {tau:.3f} / {p:.3f}")
    print(f"Tendência: {tendencia}")

    # Gráfico
    serie.plot(title="Comprometimento de renda das famílias – %", ylabel="% da renda")
    plt.tight_layout()
    arquivo = os.path.join(OUT_DIR, "comprometimento_trend.png")
    plt.savefig(arquivo)
    plt.show()
    print(f"Gráfico salvo em {arquivo}")


if __name__ == "__main__":
    main()
```

Este trabalho analisa a série “Comprometimento de renda das famílias com o serviço da dívida” (**código SGS 29265 do Banco Central do Brasil**). O objetivo é verificar se, de 2005 até hoje, a parcela da renda destinada ao pagamento de dívidas segue uma tendência crescente, estável ou decrescente.

Download dos dados diretamente da API aberta do BCB.
Como a API limita cada requisição a 10 anos, o script divide o período total em blocos e depois concatena tudo.

Pré-processamento: conversão das datas, ordenação e definição da série mensal.

Teste de tendência

Regressão linear para obter a inclinação anual (slope).

Mann-Kendall (teste não paramétrico) para confirmar estatisticamente a direção da tendência.

Classificação da tendência

Se a inclinação for positiva e o p-valor < 0,05 → Crescente.

Se a inclinação for negativa e p-valor < 0,05 → Decrescente.

Caso contrário → Estável.

Visualização: gráfico de linha exibindo toda a série, permitindo ao professor enxergar variações nos diferentes ciclos econômicos.

```python
"""
Tendências da Selic (SGS 11) e da Dívida Líquida do GG (SGS 4536)
Versão para Jupyter / Colab – exibe e salva o gráfico
"""
from datetime import datetime, timedelta
import os, io, requests
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import kendalltau
from sklearn.linear_model import LinearRegression

SERIES = {"selic": 11, "debt": 4536}
OUT_DIR = "outputs_p2"
os.makedirs(OUT_DIR, exist_ok=True)

# ------------------------------------------------ utilidades
def _fetch_chunk(series_id: int, start: str, end: str) -> pd.DataFrame:
    url = (f"https://api.bcb.gov.br/dados/serie/bcdata.sgs.{series_id}/dados"
           f"?formato=json&dataInicial={start}&dataFinal={end}")
    r = requests.get(url, timeout=30)
    r.raise_for_status()
    # evitar FutureWarning do pandas:
    df = pd.read_json(io.StringIO(r.text), dtype={"valor": "float"})
    df["data"] = pd.to_datetime(df["data"], format="%d/%m/%Y")
    return df

def fetch_series(series_id, start, end):
    s_dt, e_dt = [datetime.strptime(d, "%d/%m/%Y") for d in (start, end)]
    dfs, chunk = [], s_dt
    while chunk <= e_dt:
        try:
            end_chunk = chunk.replace(year=chunk.year + 9)
        except ValueError:
            end_chunk = chunk + timedelta(days=365*9+2)
        if end_chunk > e_dt:
            end_chunk = e_dt
        dfs.append(_fetch_chunk(series_id,
                                chunk.strftime("%d/%m/%Y"),
                                end_chunk.strftime("%d/%m/%Y")))
        chunk = end_chunk + timedelta(days=1)
    return pd.concat(dfs).sort_values("data").reset_index(drop=True)

def mann_kendall(s):           return kendalltau(range(len(s)), s)
def linear_slope(s):
    x = (s.index.astype(int)//10**9).values.reshape(-1,1)
    y = s.values.reshape(-1,1)
    coef = LinearRegression().fit(x,y).coef_[0][0]
    return coef*60*60*24*365      # por ano

def analyze(name, s, ylabel, png):
    slope = linear_slope(s)
    tau, p = mann_kendall(s)
    trend = ("CRESCENTE" if slope>0 and p<0.05
             else "DECRESCENTE" if slope<0 and p<0.05
             else "ESTÁVEL")

    print(f"\n{name}")
    print(f"  slope.............: {slope:.4f} /ano")
    print(f"  Mann-Kendall τ/p..: {tau:.3f} / {p:.3f}")
    print(f"  Tendência.........: {trend}")

    s.plot(title=name, ylabel=ylabel)
    plt.tight_layout()
    plt.savefig(png)   # grava arquivo
    plt.show()         # exibe
    plt.clf()          # limpa para o próximo gráfico

# ------------------------------------------------ execução
hoje = datetime.today().strftime("%d/%m/%Y")
selic_raw = fetch_series(SERIES["selic"], "01/01/2010", hoje)
debt_raw  = fetch_series(SERIES["debt"],  "01/01/2006", hoje)

selic_m = selic_raw.set_index("data")["valor"].resample("M").mean()   # média mensal
debt_m  = debt_raw.set_index("data")["valor"]                         # já mensal

print("=== Tendências macroeconômicas ===")
analyze("SELIC – média mensal", selic_m, "% a.a.", f"{OUT_DIR}/selic_trend.png")
analyze("Dívida líquida do GG (% PIB)", debt_m, "% PIB", f"{OUT_DIR}/debt_trend.png")

print(f"\nPNG exportados em: {OUT_DIR}/")
```

Objetivo
Investigar se existem tendências estatisticamente significativas na

Taxa Selic – média diária (SGS 11), reamostrada para média mensal, e

Dívida líquida do governo geral – % do PIB (SGS 4536), série mensal original,
entre janeiro de 2010 (Selic) / janeiro de 2006 (Dívida) e hoje.

Passo-a-passo da análise
Coleta dos dados

API aberta do BCB (bcdata.sgs.{id}) com fatias de 10 anos para contornar o limite imposto desde mar/2025.

Pré-processamento

Conversão de datas, ordenação e indexação.

Selic diária → média mensal (resample “M”).

Dívida já é mensal; nenhuma agregação extra.

Teste de tendência
Para cada série:

Regressão linear simples contra o tempo → inclinação anual (slope).

Mann-Kendall → teste não paramétrico para confirmar monotonicidade.

Critério de classificação

p < 0,05 & slope > 0 → tendência crescente.

p < 0,05 & slope < 0 → tendência decrescente.

Caso contrário → estável/indefinida.

Visualização
Dois gráficos de linha (Selic e Dívida) salvam-se em PNG e são exibidos no notebook.

Dívida líquida do GG (% PIB)	+1,4050 p.p./ano	0,423 / 0,000	Crescente (aumento estrutural, apesar do recuo pós-pandemia)

Conclusões

Dívida pública: mantém trajetória ascendente e significativa, refletindo déficits crônicos e o choque fiscal da COVID-19, mesmo após a leve reversão recente.

A comparação ilustra que política monetária (Selic) pode oscilar para conter a inflação, enquanto o endividamento do setor público segue dinâmica própria, mais lenta e estrutural.
