1 - Mineração de Dados (MD) é uma técnica utilizada para identificar padrões úteis em grandes quantidades de
dados, também é considerada uma das atividades do processo de descoberta de conhecimento. Neste sentido,
apresente exemplos sucintos em que técnicas como classificação, regressão, clusterização, associação e
detecção de anomalias possam ser utilizadas em um processo de MD (2,5)

1. Classificação
- Exemplo:
Previsão de cancelamento de clientes em uma empresa.

Utiliza dados históricos dos clientes (tempo de contrato, reclamações, uso de serviços) para classificá-los em "vai cancelar" ou "não vai cancelar".

2. Regressão
- Exemplo:
Previsão do valor de vendas de um produto nos próximos meses.

Com base em dados históricos como preço, campanhas de marketing e sazonalidade, estima-se um valor numérico.

3. Clusterização
- Exemplo:
Segmentação de clientes para campanhas de marketing.

Agrupa clientes com comportamentos semelhantes (ex: "usuários frequentes de cartão", "usuários ocasionais", "novos clientes").

4. Associação
- Exemplo:
Análise de mercado em supermercados.

Identifica regras de associação, como: “clientes que compram pão e manteiga também compram leite”.

5. Detecção de Anomalias
- Exemplo:
Identificação de transações suspeitas em cartões de crédito.

Detecta padrões fora do normal, como uma compra de alto valor no cartão, indicando possível fraude.

3 - É possível prever a acurácia com base no tempo de execução e número de iterações? Atingível
